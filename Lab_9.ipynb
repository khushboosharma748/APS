# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python

import heapq

def dijkstra(graph, start):
    # Distance dictionary
    distances = {node: float('inf') for node in graph}
    distances[start] = 0

    # Path dictionary
    previous_nodes = {node: None for node in graph}

    # Min-heap priority queue
    pq = [(0, start)]

    while pq:
        current_distance, current_node = heapq.heappop(pq)

        # Skip if we already found a shorter path
        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            if distance < distances[neighbor]:
                distances[neighbor] = distance
                previous_nodes[neighbor] = current_node
                heapq.heappush(pq, (distance, neighbor))

    return distances, previous_nodes

def shortest_path(previous_nodes, start, target):
    path = []
    current = target
    while current is not None:
        path.insert(0, current)
        current = previous_nodes[current]
    return path if path[0] == start else []

# Example graph with 2 extra nodes (E, F)
graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10},
    'C': {'A': 2, 'B': 5, 'D': 3, 'E': 4},
    'D': {'B': 10, 'C': 3, 'F': 11},
    'E': {'C': 4, 'F': 2},
    'F': {'D': 11, 'E': 2}
}

# Test with different sources
for source in ['A', 'C', 'E']:
    distances, previous_nodes = dijkstra(graph, source)
    print(f"\nSource: {source}")
    for node in graph:
        path = shortest_path(previous_nodes, source, node)
        print(f"Shortest path from {source} to {node}: {path}, Distance = {distances[node]}")
```

```python
def get_user_graph():
    """Function to build a graph from user input"""
    graph = {}
    print("Enter graph edges (source destination weight). Type 'done' to finish.")
    while True:
        user_input = input("Edge: ")
        if user_input.lower() == 'done':
            break
        try:
            u, v, w = user_input.split()
            w = float(w)
            if u not in graph: graph[u] = {}
            if v not in graph: graph[v] = {}
            graph[u][v] = w
            graph[v][u] = w # Assuming undirected graph for this example
        except ValueError:
            print("Invalid format. Use: NodeA NodeB 10")
    return graph

def run_dijkstra_interactive():
    # Using the expanded graph from your requirements (adding G and H)
    extended_graph = {
        'A': {'B': 4, 'C': 2},
        'B': {'A': 4, 'C': 5, 'D': 10},
        'C': {'A': 2, 'B': 5, 'D': 3, 'E': 4},
        'D': {'B': 10, 'C': 3, 'F': 11},
        'E': {'C': 4, 'F': 2, 'G': 3},
        'F': {'D': 11, 'E': 2, 'H': 1},
        'G': {'E': 3, 'H': 6},
        'H': {'F': 1, 'G': 6}
    }

    source_node = input(f"Enter source node (Choose from {list(extended_graph.keys())}): ").upper()

    if source_node in extended_graph:
        distances, previous_nodes = dijkstra(extended_graph, source_node)
        print(f"\nResults for Source Node: {source_node}")
        print("-" * 30)
        for target in extended_graph:
            path = shortest_path(previous_nodes, source_node, target)
            print(f"Target {target}: Path {path} | Distance: {distances[target]}")
    else:
        print("Node not found in graph.")

# Run the test with the extended graph
run_dijkstra_interactive()
```
